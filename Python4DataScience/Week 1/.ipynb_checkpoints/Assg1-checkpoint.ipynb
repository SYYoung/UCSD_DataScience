```python
stopword = set(line.strip() for line in open('stopwords'))
```

```python
punc_list = [".", ",", "\"", "“"]
```

```python
f1 = open('98-0.txt', encoding='utf8')
word_list = {}
for word in f1.read().lower().split():
    word = word.replace(".","")
    word = word.replace(",", "")
    word = word.replace("\"", "")
    word = word.replace("“", "")
    if word not in stopword:
        if word not in word_list:
            word_list[word] = 1
        else:
            word_list[word] += 1
```

```python
d = collections.Counter(word_list)
d.most_common(10)
```
